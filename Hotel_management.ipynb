```python
import numpy as np

rooms = np.zeros(9)   # 0 = available, 1 = booked
customers = {}
payments = []


# Room layout with bed type
def room_layout():

    print("\nROOM LAYOUT")

    for i in range(9):

        if i < 3:
            room_type = "2-Bed"
        elif i < 6:
            room_type = "3-Bed"
        else:
            room_type = "4-Bed"

        if rooms[i] == 0:
            status = "Available"
        else:
            status = "Booked"

        print("Room", i+1, "-", room_type, "-", status)


# Booking room
def book_room():

    room_layout()

    room = int(input("Enter room number: "))

    if rooms[room-1] == 0:

        name = input("Enter customer name: ")
        phone = input("Enter phone number: ")

        rooms[room-1] = 1
        customers[room] = (name, phone)

        print("Room booked successfully")

    else:
        print("Room already booked")


# Check-in
def check_in():

    room = int(input("Enter room number: "))

    if rooms[room-1] == 1:

        name, phone = customers[room]

        print("Customer Name:", name)
        print("Phone:", phone)
        print("Check-in successful")

    else:
        print("Room not booked")


# Check-out
def check_out():

    room = int(input("Enter room number: "))

    if rooms[room-1] == 1:

        days = int(input("Enter number of days: "))
        price = 1000

        bill = days * price

        payments.append(bill)

        name, phone = customers[room]

        print("Customer:", name)
        print("Total Bill:", bill)

        rooms[room-1] = 0
        del customers[room]

    else:
        print("Room already empty")


# Payment summary
def payment_summary():

    if len(payments) == 0:
        print("No payments yet")

    else:
        print("Payments:", payments)
        print("Total earnings:", sum(payments))
        print("Highest payment:", max(payments))


# Main menu
while True:

    print("\n--- HOTEL MANAGEMENT ---")
    print("1. Show Room Layout")
    print("2. Book Room")
    print("3. Check-In")
    print("4. Check-Out")
    print("5. Payment Summary")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        room_layout()

    elif choice == 2:
        book_room()

    elif choice == 3:
        check_in()

    elif choice == 4:
        check_out()

    elif choice == 5:
        payment_summary()

    elif choice == 6:
        break

    else:
        print("Invalid choice")
```
